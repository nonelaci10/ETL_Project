```python
import pandas as pd
from sqlalchemy import create_engine
from config import pw
```

```python
fm_data = "Resources/Export.csv"
fm_df = pd.read_csv(fm_data)
fm_df.head()
```

```python
pop_data = "Resources/sub-est2018_all.csv"
pop_df = pd.read_csv(pop_data, encoding = "ISO-8859-1")
pop_df.head()
```

```python
farmers_df = fm_df[['FMID','MarketName','city','State','y','x']]
farmers_df = farmers_df.rename(columns={'MarketName': 'Market_Name','city':'City','x':'Lon','y':'Lat'})
farmers_df.head()
```

```python
population_df = pop_df[['NAME','STNAME','POPESTIMATE2018']]
population_df.head()
```

```python
population_df["NAME"]= population_df["NAME"].str.replace(" city", "", case = True)
population_df.head()
```

```python
population_df["NAME"]= population_df["NAME"].str.replace(" town", "", case = True)
population_df.head()
```

```python
population_df["NAME"]= population_df["NAME"].str.replace(" village", "", case = True)
population_df.head()
```

```python
population_df = population_df.rename(columns={'NAME':'City', 'STNAME':'State', 'POPESTIMATE2018':'Population'})
population_df.head()
```

```python
#Creating Database Connection

connection_string = "postgres:config@localhost:5432/ETL"
engine = create_engine(f'postgresql://{connection_string}')
```
